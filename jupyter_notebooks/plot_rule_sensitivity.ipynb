```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("../output_data/benchmark_rule_sensitivity_lymph.csv")
```

```python
df.head(10)
```

```python
labels = {
    "SLS": "pyIDS - SLS",
    "DLS": "pyIDS - DLS",
    "DUSM": "pyIDS - DUSM",
    "RUSM": "pyIDS - RUSM",
    "pyARC - M1": "pyARC - M1",
    "pyARC - M2": "pyARC - M2"
}
```

```python
algorithm_dict = {
    "DLS": "blue",
    "SLS": "green",
    "DUSM": "orange",
    "RUSM": "red",
    "pyARC - M1": "black",
    "pyARC - M2": "cyan"
}

plt.figure()

for algorithm, color in algorithm_dict.items():
    df_alg = df[df["algorithm"] == algorithm]
    
    label = labels[algorithm]
    
    plt.plot(df_alg["rule_num"], df_alg["duration"], label=label, color=color)

plt.legend()
plt.xlabel("Počet vstupních pravidel")
plt.ylabel("Čas běhu algoritmu [s]")
```
