```python
import pandas as pd
import numpy as np
from pyids.data_structures import IDS, mine_CARs, IDSRuleSet

from pyarc.qcba.data_structures import QuantitativeDataFrame

import random 



df = pd.read_csv("../data/iris0.csv")
cars = mine_CARs(df, 80, sample=False)
ids_ruleset = IDSRuleSet.from_cba_rules(cars).ruleset

quant_dataframe = QuantitativeDataFrame(df)

for r in reversed(sorted(cars)):
    print(r)


def train_ids():
    ids = IDS(algorithm="SLS")
    ids.fit(class_association_rules=cars, algorithm="DLS", quant_dataframe=quant_dataframe, debug=False, random_seed=None, lambda_array=[1, 1, 1, 1, 1, 1, 1])
    auc = ids.score(quant_dataframe)
```

```python
cProfile.run("train_ids()")
```
