```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("../output_data/benchmark_optimization_algorithms.csv")
```

```python
df
```

```python
for algorithm, color in dict(DLS="blue", SLS="green", DUSM="orange", RUSM="red").items():
    df_alg = df[df["algorithm"] == algorithm]
    
    plt.plot(df_alg["rule_num"], df_alg["duration"], label=algorithm, color=color)

plt.legend()
plt.xlabel("Počet vstupních pravidel")
plt.ylabel("Čas optimalizace [s]")
```
