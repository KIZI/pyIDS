```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("../output_data/benchmark_datasize_sensitivity_lymph.csv")
```

```python
df.head(10)
```

```python
algorithm_dict = {
    "DLS": "blue",
    "SLS": "green",
    "DUSM": "orange",
    "RUSM": "red",
    "pyARC - M1": "black",
    "pyARC - M2": "cyan"
}

for algorithm, color in algorithm_dict.items():
    df_alg = df[df["algorithm"] == algorithm]
    
    plt.plot(df_alg["data_count"], df_alg["duration"], label=algorithm, color=color)

plt.legend()
plt.xlabel("Počet vstupních řádků dat")
plt.ylabel("Čas optimalizace [s]")
```
